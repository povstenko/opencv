# Translation

Translation is the shifting of an image along the x and y axis. Using translation, we can shift an image up, down, left, or right, along with any combination of the above.

### import the necessary packages

```python
from matplotlib import pyplot as plt
import numpy as np
import imutils
import cv2
```

### load the image and show it

```python
image = cv2.imread("image/img.jpg")
plt.imshow(image)
plt.show()
```

```python
# NOTE: Translating (shifting) an image is given by a NumPy matrix in
# the form:
#	[[1, 0, shiftX], [0, 1, shiftY]]
# You simply need to specify how many pixels you want to shift the image
# in the X and Y direction -- let's translate the image 25 pixels to the
# right and 50 pixels down
M = np.float32([[1, 0, 25], [0, 1, 50]])

shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt.imshow(shifted)
plt.show()
```

```python
# now, let's shift the image 50 pixels to the left and 90 pixels up, we
# accomplish this using negative values
M = np.float32([[1, 0, -50], [0, 1, -90]])

shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt.imshow(shifted)
plt.show()
```

### finally, let's use our helper function in imutils to shift the image down 100 pixels

```python
shifted = imutils.translate(image, 0, 100)
plt.imshow(shifted)
plt.show()
```
