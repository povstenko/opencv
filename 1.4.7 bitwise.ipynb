# Bitwise operations

Bitwise operations operate in a binary manner and are represented as grayscale images. A given pixel is turned “off” if it has a value of zero and it is turned “on” if the pixel has a value greater than zero.

### import the necessary packages

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### draw a rectangle

```python
rectangle = np.zeros((300, 300), dtype = "uint8")
cv2.rectangle(rectangle, (25, 25), (275, 275), 255, -1)

plt.imshow(rectangle)
plt.title("Rectangle")
plt.show()
```

### draw a circle

```python
circle = np.zeros((300, 300), dtype = "uint8")
cv2.circle(circle, (150, 150), 150, 255, -1)

plt.imshow(circle)
plt.title("Circle")
plt.show()
```

### AND

A bitwise 'AND' is only True when both rectangle and circle have a value that is 'ON.' Simply put, the bitwise AND function examines every pixel in rectangle and circle. If both pixels have a value greater than zero, that pixel is turned 'ON' (i.e set to 255 in the output image). If both pixels are not greater than zero, then the output pixel is left 'OFF' with a value of 0.

```python
bitwiseAnd = cv2.bitwise_and(rectangle, circle)

plt.imshow(bitwiseAnd)
plt.title("AND")
plt.show()
```

### OR

A bitwise 'OR' examines every pixel in rectangle and circle. If EITHER pixel in rectangle or circle is greater than zero, then the output pixel has a value of 255, otherwise it is 0.

```python
bitwiseOr = cv2.bitwise_or(rectangle, circle)

plt.imshow(bitwiseOr)
plt.title("OR")
plt.show()
```

### XOR

The bitwise 'XOR' is identical to the 'OR' function, with one exception: both rectangle and circle are not allowed to BOTH have values greater than 0.

```python
bitwiseXor = cv2.bitwise_xor(rectangle, circle)

plt.imshow(bitwiseXor)
plt.title("XOR")
plt.show()
```

### NOT

Finally, the bitwise 'NOT' inverts the values of the pixels. Pixels with a value of 255 become 0, and pixels with a value of 0 become 255.

```python
bitwiseNot = cv2.bitwise_not(circle)

plt.imshow(bitwiseNot)
plt.title("NOT")
plt.show()
```
