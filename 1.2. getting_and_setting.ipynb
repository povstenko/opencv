# Image basics

### import the necessary packages

```python
from matplotlib import pyplot as plt
import cv2
```

### load the image, grab its dimensions, and show it

```python
image = cv2.imread("image/dsc_logo.jpg")
(h, w) = image.shape[:2]
plt.imshow(image)
plt.show()
```

```python
print("Height:", h)
print("Width:", w)
```

## Accessing and Manipulating Pixels

```python
# images are just NumPy arrays. The top-left pixel can be found at (0, 0)
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {r}, Green: {g}, Blue: {b}".format(r=r, g=g, b=b))
 
# now, let's change the value of the pixel at (0, 0) and make it red
image[0, 0] = (0, 0, 255)
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {r}, Green: {g}, Blue: {b}".format(r=r, g=g, b=b))
```

### compute the center of the image

```python
# it is simply the width and height divided by two
(cX, cY) = (w // 2, h // 2)
```

### let's grab the top-left corner

```python
# since we are using NumPy arrays, we can apply slicing and grab large chunks of the image
tl = image[0:cY, 0:cX]
plt.imshow(tl)
plt.show()
```

### in a similar fashion, let's grab the top-right, bottom-right, and bottom-left corners and display them

```python
tr = image[0:cY, cX:w]
br = image[cY:h, cX:w]
bl = image[cY:h, 0:cX]
plt.imshow(tr)
plt.show()
plt.imshow(br)
plt.show()
plt.imshow(bl)
plt.show()
```

### now let's make the top-left corner of the original image green

```python
image[0:cY, 0:cX] = (0, 255, 0)
 
# Show our updated image
plt.imshow(image)
plt.show()
```
