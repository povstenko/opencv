# Masking

Bitwise operations operate in a binary manner and are represented as grayscale images. A given pixel is turned “off” if it has a value of zero and it is turned “on” if the pixel has a value greater than zero.

### import the necessary packages

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### load the image and display it

```python
image = cv2.imread("image/house.jpg")

plt.imshow(image)
plt.title("Original")
plt.show()
```

### masking

Masking allows us to focus only on parts of an image that interest us. A mask is the same size as our image, but has only two pixel values, 0 and 255. Pixels with a value of 0 are ignored in the orignal image, and mask pixels with a value of 255 are allowed to be kept.

```python
mask = np.zeros(image.shape[:2], dtype="uint8")
cv2.rectangle(mask, (1500, 1200), (800, 700), 255, -1)

plt.imshow(mask)
plt.title("Mask")
plt.show()
```

### Apply our mask

```python
masked = cv2.bitwise_and(image, image, mask=mask)

plt.imshow(masked)
plt.title("Mask Applied to Image")
plt.show()
```

### Circular mask

```python
mask = np.zeros(image.shape[:2], dtype="uint8")
cv2.circle(mask, (1100, 900), 300, 255, -1)
cv2.circle(mask, (1100, 900), 290, 0, -1)
cv2.circle(mask, (1100, 900), 280, 255, -1)
masked = cv2.bitwise_and(image, image, mask=mask)

plt.imshow(mask)
plt.title("Mask")
plt.show()

plt.imshow(masked)
plt.title("Mask Applied to Image")
plt.show()
```
