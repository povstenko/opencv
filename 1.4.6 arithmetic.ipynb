# Image arithmetic

When we crop an image, we want to remove the outer parts of the image that we are not interested in. This is commonly called selecting our Region of Interest, or more simply, our ROI.

### import the necessary packages

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### load the image and show it

```python
image = cv2.imread("image/img.jpg")
plt.imshow(image)
plt.title("Original")
plt.show()
```

### Arithmetics

images are NumPy arrays, stored as unsigned 8 bit integers -- this implies that the values of our pixels will be in the range [0, 255]; when using functions like cv2.add and cv2.subtract, values will be clipped to this range, even if the added or subtracted values fall outside the range of [0, 255]. Check out an example:

```python
print("max of 255: {}".format(str(cv2.add(np.uint8([200]), np.uint8([100])))))
print("min of 0: {}".format(str(cv2.subtract(np.uint8([50]), np.uint8([100])))))
```

NOTE: if you use NumPy arithmetic operations on these arrays, the value will be modulo (wrap around) instead of being  clipped to the [0, 255] range. This is important to keep in mind when working with images.

```python
print("wrap around: {}".format(str(np.uint8([200]) + np.uint8([100]))))
print("wrap around: {}".format(str(np.uint8([50]) - np.uint8([100]))))
```

### Increase the intensity

let's increase the intensity of all pixels in our image by 100 -- we accomplish this by constructing a NumPy array that is the same size of our matrix (filled with ones) and the multiplying it by 100 to create an array filled with 100's, then we simply add the images together; notice how the image is "brighter"

```python
M = np.ones(image.shape, dtype = "uint8") * 100
added = cv2.add(image, M)

plt.imshow(added)
plt.title("Added")
plt.show()
```

### Decrease the intensity

similarly, we can subtract 50 from all pixels in our image and make it darker

```python
M = np.ones(image.shape, dtype = "uint8") * 50
subtracted = cv2.subtract(image, M)

plt.imshow(subtracted)
plt.title("Subtracted")
plt.show()
```
