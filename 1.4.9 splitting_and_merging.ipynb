# Splitting and merging channels

A color image consists of multiple channels: a Red, a Green, and a Blue component. We have seen that we can access these components via indexing into NumPy arrays. But what if we wanted to split an image into its respective components?

### import the necessary packages

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### load the image and display channels

```python
image = cv2.imread("image/img.jpg")

plt.imshow(image)
plt.title("Original")
plt.show()
```

### Split channels

Load the image and grab each channel: Red, Green, and Blue. It's important to note that OpenCV stores an image as NumPy array with its channels in reverse order! When we call cv2.split, we are actually getting the channels as Blue, Green, Red!

```python

(B, G, R) = cv2.split(image)

plt.imshow(R)
plt.title("Red")
plt.show()

plt.imshow(G)
plt.title("Green")
plt.show()

plt.imshow(B)
plt.title("Blue")
plt.show()
```

### Merge channels

merge the image back together again

```python
merged = cv2.merge([B, G, R])

plt.imshow(merged)
plt.title("Merged")
plt.show()
```

### visualize each channel in color

```python
zeros = np.zeros(image.shape[:2], dtype = "uint8")

plt.imshow(cv2.merge([R, zeros, zeros]))
plt.title("Red")
plt.show()

plt.imshow(cv2.merge([zeros, G, zeros]))
plt.title("Green")
plt.show()

plt.imshow(cv2.merge([zeros, zeros, B]))
plt.title("Blue")
plt.show()
```
